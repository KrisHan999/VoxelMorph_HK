```python
import numpy as np
import glob
import matplotlib.pyplot as plt
```

```python
test_affine_names = glob.glob('./data/test_affine/*.npz')
```

```python
print(test_affine_names)
```

```python
test_affine_movings = []
test_affine_fixeds = []
test_affine_segs = []
for i in range(len(test_affine_names)):
    test_affine_movings.append(np.load(test_affine_names[1])['moving'])
    test_affine_fixeds.append(np.load(test_affine_names[1])['fixed'])
    test_affine_segs.append(np.load(test_affine_names[1])['seg'])
```

```python
def print_image(vol):
    depth, height, width = vol.shape
    fig, ((ax1, ax2, ax3),(ax4, ax5, ax6)) = plt.subplots(2, 3, figsize=(20, 10))
    ax1.imshow(np.mean(vol[:, :, :], axis=0))
    ax2.imshow(np.mean(vol[:, :, :], axis=1))
    ax3.imshow(np.mean(vol[:, :, :], axis=2))
    ax4.imshow(vol[depth//2, :, :])
    ax5.imshow(vol[:, height//2, :])
    ax6.imshow(vol[:, :, width//2])
```

```python
print_image(test_affine_segs[0])
```

```python
print_image(test_affine_segs[1])
```

```python
atlas = np.load('./data/atlas_norm.npz')
atlas_seg = atlas['seg']
print_image(atlas_seg)
```
